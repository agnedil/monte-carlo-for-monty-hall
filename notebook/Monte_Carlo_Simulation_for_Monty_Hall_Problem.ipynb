# Monte Carlo Simulation for [Monty Hall Problem](https://en.wikipedia.org/wiki/Monty_Hall_problem)

## Using random

```python
import random

def generate_triplets(n):
    triplets = []
    for _ in range(n):
        triplet = [0, 0, 1]
        random.shuffle(triplet)
        triplets.append(triplet)
    return triplets


def pick_from_triplets(triplets):
    result_stay = []
    result_change = []
    picks = []
    for triplet in triplets:
        # Step 1: Randomly pick one index
        first_pick = random.randint(0, 2)
        result_stay.append(triplet[first_pick])

        # Step 2: Find and pick an index with a 0, different from the first pick
        zero_indices = [i for i, value in enumerate(triplet) if value == 0 and i != first_pick]
        if not zero_indices:          # If the first pick was a 0, pick the other 0
            zero_indices = [i for i, value in enumerate(triplet) if value == 0]
        second_pick = random.choice(zero_indices)

        # Step 3: Pick the third, untouched index
        third_pick = 3 - first_pick - second_pick
        result_change.append(triplet[third_pick])
        picks.append([first_pick, second_pick, third_pick])

    return result_stay, result_change, picks


# Generate n triplets
n = 100000
triplets = generate_triplets(n)

result_stay, result_change, picks = pick_from_triplets(triplets)

print(f'Stay: {sum(result_stay)}    Change: {sum(result_change)}')
set([sum(i) for i in picks])
```

## Using numpy

```python
import numpy as np

def generate_triplets(n):
    # Generate an array of shape (n, 3) filled with zeros and then set one 1 per row
    triplets = np.zeros((n, 3), dtype=int)
    ones_indices = np.random.randint(0, 3, size=n)
    triplets[np.arange(n), ones_indices] = 1
    return triplets

def pick_from_triplets(triplets):
    n = triplets.shape[0]
    result_stay = []
    result_change = []
    picks = []

    for i in range(n):
        triplet = triplets[i]

        # Step 1: Randomly pick one index
        first_pick = np.random.randint(0, 3)
        result_stay.append(triplet[first_pick])

        # Step 2: Find and pick an index with a 0, different from the first pick
        zero_indices = [j for j in range(3) if triplet[j] == 0 and j != first_pick]
        if not zero_indices:  # If the first pick was a 0, pick the other 0
            zero_indices = [j for j in range(3) if triplet[j] == 0]
        second_pick = np.random.choice(zero_indices)

        # Step 3: Pick the third, untouched index
        third_pick = 3 - first_pick - second_pick
        result_change.append(triplet[third_pick])
        picks.append([first_pick, second_pick, third_pick])

    return result_stay, result_change, picks


# Generate n triplets
n = 100000
triplets = generate_triplets(n)

result_stay, result_change, picks = pick_from_triplets(triplets)
print(f'Stay: {sum(result_stay)}    Change: {sum(result_change)}')
set([sum(i) for i in picks])
```
